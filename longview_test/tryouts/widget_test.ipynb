```python
import ipywidgets as widgets

class PrinterX:
    def __init__(self):
        self.w = w=widgets.HTML()
    def show(self):
        return self.w
    def write(self,s):
        self.w.value = s
```

```python
p=PrinterX()
p.show()
```

```python
p.write("ffffff")
```
