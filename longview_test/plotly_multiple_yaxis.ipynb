```python
import plotly
import plotly.colors
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
def get_pallet_color(i):
    return plotly.colors.DEFAULT_PLOTLY_COLORS[i % len(plotly.colors.DEFAULT_PLOTLY_COLORS)]
```

```python
trace1 = go.Scatter(
    x=[1, 2, 3],
    y=[4, 5, 6],
    name='yaxis1 data',
    line=dict(color=get_pallet_color(0))
    
)
trace2 = go.Scatter(
    x=[5, 6, 7],
    y=[400000, 500000, 600000],
    name='yaxis2 data',
    yaxis='y2',
    line=dict(color=get_pallet_color(1))
)
trace3 = go.Scatter(
    x=[5, 6, 7],
    y=[400, 100, 50],
    name='yaxis3 data',
    yaxis='y3',
    line=dict(color=get_pallet_color(2))
)
data = [trace1, trace2, trace3]
layout = go.Layout(
    title='multiple y-axes example',
    xaxis=dict(
        domain=[0, 0.75]
    ),
    yaxis2=dict(
        title='yaxis1 title',
        side='right',
        overlaying='y',
        titlefont=dict(color=get_pallet_color(1)),
        tickfont=dict(color=get_pallet_color(1))
    ),
    yaxis3=dict(
        title='yaxis2 title',
        anchor='free',
        overlaying='y',
        side='right',
        position=0.85,
        titlefont=dict(color=get_pallet_color(2)),
        tickfont=dict(color=get_pallet_color(2))
    )
)
fig = go.FigureWidget(data=data, layout=layout)
fig
```
