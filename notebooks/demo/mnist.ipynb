```python
import tensorwatch as tw
```

```python
trl = tw.open('lambda d:(d.x, d.metrics.batch_loss)', event_name='batch', type='mpl-line', 
              xtitle='Epoch', ytitle='Train Loss', clear_after_end=False)
trl.show()
```

```python
tra = tw.open('lambda d:(d.x, d.metrics.batch_accuracy)', event_name='batch', type='mpl-line', 
              ytitle='Train Accuracy', clear_after_end=False, vis=trl)
```

```python
tsl = tw.open('lambda d:(d.x, d.metrics.batch_loss)', event_name='batch', type='mpl-line', 
              xtitle='Epoch', ytitle='Test Loss', clear_after_end=False, cli_id=1, vis=trl)
```

```python
tsl = tw.open('lambda d:(d.x, d.metrics.batch_accuracy)', event_name='batch', type='mpl-line', 
              xtitle='Epoch', ytitle='Test Accuracy', clear_after_end=False, cli_id=1, vis=trl)
```

```python
grads = tw.open('lambda d:agg_params(d.model, lambda p: p.grad.abs().mean().item())', event_name='batch', throttle=1,
               title="Weight Gradients",  xtitle='Epoch', ytitle='Gradients', clear_after_each=True, type='mpl-line',
               clear_after_end=False, history_len=20)
grads.show()
```
